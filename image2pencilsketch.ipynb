**importing the libraries**

```python
from google.colab.patches import cv2
from google.colab.patches import cv2_imshow
```

**reading the rgb image file**

```python
i=cv2.imread("/content/krishna.jpg")
cv2_imshow(i)
```

converting the rgb image to grey_scale image

```python
grey=cv2.cvtColor(i,cv2.COLOR_BGR2GRAY)
cv2_imshow(grey)
```

**Making the image inverted**

```python
inverted=255-grey
cv2_imshow(inverted)
```

making the image blur

```python
blur=cv2.GaussianBlur(inverted,(21,21),0)
```

blurred image to pencil sketch

```python
inverted_blur=255-blur
ps=cv2.divide(grey,inverted_blur,scale=256)
cv2_imshow(ps)
```

## **Making the model**

```python
def image2pencilsketch(a):
  I=cv2.imread(a)
  print("\n\n")
  I=cv2.resize(I,(512,512))
  cv2_imshow(I)
  grey=cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)
  inverted=255-grey
  blur=cv2.GaussianBlur(inverted,(21,21),0)
  inverted_blur=255-blur
  ps=cv2.divide(grey,inverted_blur,scale=256)
  cv2_imshow(ps)
```

### ***Deploying the model***

```python
while(True):
  print("enter 1 to convert the image \n enter 0 to exit")
  k=input("")
  if k=='1':
    image=str(input("enter the image path:"))
    print("\n")
    image2pencilsketch(image)
  elif k=='0':
    break
  else:
    print("invalid input\n")
```
